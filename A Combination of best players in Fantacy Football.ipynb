## Problem Statement
#### I have dataset of 147 players of fantcy football team and they wants to choose best 9 player among them for upcoming match.
For Selecting 9 best player they wants __combination__ of players with some __conditions__, which are below
<br>**1** Player Must be __9__.
<br>**2** The Pos (Type) of player must be following: __1 QB, 2 RB, 1 DST, 1 TE, 3 WR and 1 (RB / WR / TE)__ number of players.
<br>**3.** Patterns for Pos can be anything but which Pattern you choose they __must be following for every combination__,
for Ex pattern could be like this: (QB, RB, RB, WR, WR, WR, TE, RB/WR/TE, DST).
<br>**4.** __Total Salary__ of the combined players must be range between __49400 to 50000__.
<br>**5.** Add up __Proj.Pts__ of combined players.
<br>**6.** Maximum __limits__ of combination is __1000__.


## Solutions

Here i solve this problem using **pandas** and **combinations of itertools** library but challanging part here was they wants **same pattern of all combinations**
<br>Which i solved using my smart logics and with help of core python concept and it take around **85 second** to create **1000** combinaiton with satisfying all the conditions.



```python
import pandas as pd
```

```python
main_df = pd.read_excel("inputFile.xlsm")
main_df.head()
```

```python
# get only columns which required
df1 = main_df[main_df.columns[:6]].copy()
df1.head()
```

```python
df1.dropna(inplace=True)
```

```python
df1.shape
```

```python
# rename column name for making easy to access any columns

df1.columns = df1.columns.str.replace(".", "")
df1.columns = df1.columns.str.replace(" ", "_")
```

```python
df1.head()
```

```python
df1.index = df1.Name
```

```python
from itertools import combinations
```

```python
combo = combinations(df1.index, 9)
```

```python
## below this is real script
```

```python
import time

startTime = time.time()
count = 0

filtered_data_list = []
for c in combo:
    if count == 1000:
        break
    
    temp = df1.loc[c , :]
    pos_lst = temp.Pos.to_list()
    if (pos_lst.count("DST") == 1 and pos_lst.count("QB") == 1 and pos_lst.count("RB") == 3 and pos_lst.count("WR") == 3 and pos_lst.count("TE") == 1) or (pos_lst.count("DST") == 1 and pos_lst.count("QB") == 1 and pos_lst.count("RB") == 2 and pos_lst.count("WR") == 4 and pos_lst.count("TE") == 1) or (pos_lst.count("DST") == 1 and pos_lst.count("QB") == 1 and pos_lst.count("RB") == 2 and pos_lst.count("WR") == 3 and pos_lst.count("TE") == 2):

#         print("pattern is match at ", count)
        sal_sum = temp.Sal.sum()
        if sal_sum > 49400 and sal_sum < 50000:
            col_dict = {"pts": "","sal": sal_sum , "QB": "", "RB1": "", "RB2": "", "WR1": "", "WR2": "", "WR3": "", "TE1": "", "MIX": "", "DST": ""}
            col_dict["pts"] = temp.Proj_Pts.sum()
            RB = []
            WR = []
            TE = []
            for i in temp.index.to_list():
                if temp.loc[i, "Pos"] == "WR":
                    WR.append(i)
                elif temp.loc[i, "Pos"] == "RB":
                    RB.append(i)
                elif temp.loc[i, "Pos"] == "TE":
                    TE.append(i)
                elif temp.loc[i, "Pos"] == "QB":
                    col_dict["QB"] = i
                elif temp.loc[i, "Pos"] == "DST":
                    col_dict["DST"] = i
            
            if len(RB) == 3:
                col_dict["MIX"] = RB.pop()
            for r in range(2):
                col_dict["RB"+str(r+1)] = RB[r]
            
            if len(WR) == 4:
                    col_dict["MIX"] = WR.pop()
            for w in range(3):
                col_dict["WR"+str(w+1)] = WR[w]
            
            if len(TE) == 2:
                col_dict["MIX"] = TE.pop()
            
            col_dict["TE1"] = TE[0]
                
            count +=1
            
            filtered_data_list.append(col_dict)

print("it takes --- %s seconds ---" % (time.time() - startTime))
```

```python
finalDf = pd.DataFrame(filtered_data_list)
finalDf.columns = ["Proj Pts","Salary" , "QB", "RB", "RB", "WR", "WR", "WR", "TE", "RB/WR/TE", "DST"]
finalDf
```
